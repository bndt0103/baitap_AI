```python
import random
GOAL_STATE = [1, 2, 3, 4, 5, 6, 7, 8, 0]
```

```python
def get_misplaced_count(state):
    return sum(1 for i in range(9) if state[i] != 0 and state[i] != GOAL_STATE[i])

def get_valid_moves(empty_idx):
    moves = {}
    row, col = divmod(empty_idx, 3)
    if row > 0: moves['UP']    = empty_idx - 3
    if row < 2: moves['DOWN']  = empty_idx + 3
    if col > 0: moves['LEFT']  = empty_idx - 1
    if col < 2: moves['RIGHT'] = empty_idx + 1
    return moves
```

```python
def RULE_MATCH(state):
    if state == GOAL_STATE:
        return "STOP"
    empty_idx = state.index(0)
    valid_moves = get_valid_moves(empty_idx)
    best_action = None
    min_misplaced = 999

    for action, new_idx in valid_moves.items():
        new_state = state[:]
        new_state[empty_idx], new_state[new_idx] = new_state[new_idx], new_state[empty_idx]
        score = get_misplaced_count(new_state)
        if score < min_misplaced or (score == min_misplaced and random.choice([True, False])):
            min_misplaced = score
            best_action = action
    return best_action

def SIMPLE_REFLEX_AGENT(percept):
    state = percept[:] 
    action = RULE_MATCH(state)
    return action   
```

```python
def print_board(state):
    for i in range(0, 9, 3):
        print(" ", state[i:i+3])
    print("-" * 15)

def run_eight_puzzle(initial_state):
    current_state = initial_state[:]
    step = 0
    print("TRẠNG THÁI BAN ĐẦU:")
    print_board(current_state)

    while step < 100:
        step += 1
        action = SIMPLE_REFLEX_AGENT(current_state)
        
        if action == "STOP":
            print(f"THÀNH CÔNG! Trạng thái đích được giải quyết sau {step - 1} bước.")
            return
        print(f"Bước {step} -> Agent di chuyển ô trống (0) sang: {action}")
        empty_idx = current_state.index(0)
        new_idx = get_valid_moves(empty_idx)[action]
        current_state[empty_idx], current_state[new_idx] = current_state[new_idx], current_state[empty_idx]
        print_board(current_state)

    print("DỪNG CHƯƠNG TRÌNH: Vượt quá giới hạn 100 bước.")
    print("Agent không tìm được đường tới đích hoặc đã bị kẹt trong vòng lặp.")
```

```python
initial = [1, 2, 3, 
           4, 5, 6, 
           0, 7, 8]
run_eight_puzzle(initial)
```
