```python
import random

# Ma trận mxn: 
# 0 -> không bụi
# 1 -> có bụi
# -1 -> vật cản (MỚI)
# 2 -> vị trí robot (hiển thị khi in)

floor = [
    [0, 1, -1],
    [1, -1, 0],
    [0, 1, 1]
]
row = len(floor)
col = len(floor[0])

while True:
    vaccum_x = random.randint(0, row - 1)
    vaccum_y = random.randint(0, col - 1)
    if floor[vaccum_x][vaccum_y] != -1:
        break
```

```python
def show_floor():
    for i in range(row):
        for j in range(col):
            if i == vaccum_x and j == vaccum_y:
                print(2, end=' ')
            elif floor[i][j] == -1:
                print('X', end=' ')
            else:
                print(floor[i][j], end=' ')
        print()
```

```python
def has_dirty():
    for row in floor:
        if 1 in row:
            return True
    return False
```

```python
def rule_match():
    global vaccum_x, vaccum_y
    state = floor[vaccum_x][vaccum_y]
    
    if state == 1:
        print("Place is dirty!")
        floor[vaccum_x][vaccum_y] = 0
        print("Action: Suck!")

    elif state == 0:
        print("Place is cleaned!")
        direction =[]
        if vaccum_x > 0 and floor[vaccum_x - 1][vaccum_y] != -1:
            direction.append((-1, 0, "UP"))
        if vaccum_x < row - 1 and floor[vaccum_x + 1][vaccum_y] != -1:
            direction.append((1, 0, "DOWN"))
        if vaccum_y > 0 and floor[vaccum_x][vaccum_y - 1] != -1:
            direction.append((0, -1, "LEFT"))
        if vaccum_y < col - 1 and floor[vaccum_x][vaccum_y + 1] != -1:
            direction.append((0, 1, "RIGHT"))

        if direction:
            move = random.choice(direction)
            dx, dy, direction_name = move
            vaccum_x += dx
            vaccum_y += dy
            print("Action: Move", direction_name)
        else:
            print("Vaccum is blocked by obstacal!")
```

```python
step = 1
while has_dirty() and step < 100:
    print(f"\n=====================Step {step} ==================")
    show_floor()
    rule_match()
    step += 1

print("===============Result=============")
if not has_dirty():
    print("All place is cleaned by vaccum!")
else:
    print("Terminal due to over step permission!")
show_floor()
```
