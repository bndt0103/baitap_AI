```python
import copy
START_STATE = [
    [1, 0, 3],
    [4, 2, 6],
    [7, 5, 8] 
]

GOAL_STATE = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]
```

```python
class Node:
    def __init__(self, state, parent=None, action=None, path_cost=0):
        self.state = state          
        self.parent = parent        
        self.action = action        
        self.path_cost = path_cost 
```

```python
def print_state(state):
    for row in state:
        print(row)
    print()

def find_blank(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j
    return -1, -1

def ACTIONS(state):
    actions = []
    r, c = find_blank(state)
    if r > 0: actions.append("Up")
    if r < 2: actions.append("Down")
    if c > 0: actions.append("Left")
    if c < 2: actions.append("Right")
    return actions
```

```python
def CHILD_NODE(node, action):
    new_state = copy.deepcopy(node.state)
    r, c = find_blank(new_state)
    if action == "Up":
        new_state[r][c], new_state[r-1][c] = new_state[r-1][c], new_state[r][c]
    elif action == "Down":
        new_state[r][c], new_state[r+1][c] = new_state[r+1][c], new_state[r][c]
    elif action == "Left":
        new_state[r][c], new_state[r][c-1] = new_state[r][c-1], new_state[r][c]
    elif action == "Right":
        new_state[r][c], new_state[r][c+1] = new_state[r][c+1], new_state[r][c]
        
    return Node(state=new_state, parent=node, action=action, path_cost=node.path_cost + 1)
```

```python
def GOAL_TEST(state):
    return state == GOAL_STATE

def in_frontier(frontier, state):
    for n in frontier:
        if n.state == state:
            return True
    return False
```

```python
def BREADTH_FIRST_SEARCH():
    node = Node(START_STATE)
    if GOAL_TEST(node.state):
        return node
        
    frontier = []
    frontier.append(node)
    reached = []
    
    while len(frontier) > 0:
        node = frontier.pop(0)
        if node.state not in reached:
            reached.append(node.state)
        if GOAL_TEST(node.state):
            return node
        for action in ACTIONS(node.state):
            child = CHILD_NODE(node, action)
            if (child.state not in reached) and not in_frontier(frontier, child.state):
                frontier.append(child)
    return None
```

```python
class ModelBasedReflexAgent:
    def __init__(self, plan):
        self.state = None        
        self.model = "8-Puzzle"  
        self.rules = plan        
        self.action = None       

    def UPDATE_STATE(self, current_state, action, percept, model):
        return percept

    def RULE_MATCH(self, state, rules):
        if len(rules) > 0:
            return rules.pop(0)
        return None

    def execute(self, percept):
        self.state = self.UPDATE_STATE(self.state, self.action, percept, self.model)
        next_action = self.RULE_MATCH(self.state, self.rules)
        self.action = next_action
        return self.action
```

```python

if __name__ == "__main__":
    solution_node = BREADTH_FIRST_SEARCH()
    if solution_node:
        plan = []
        curr = solution_node
        while curr.parent is not None:
            plan.append(curr.action)
            curr = curr.parent
        plan.reverse() 
        print(f"Have found the way. {len(plan)} step: {plan}\n")

        agent = ModelBasedReflexAgent(plan)
        current_percept = START_STATE
        print("STATE BEGIN")
        print_state(current_percept)
        step = 1
        while True:
            action = agent.execute(current_percept)
            if action is None:
                print("Agent complete the state goal!")
                break
            print(f"Step {step}: Agent move: {action}")
            temp_node = Node(current_percept)
            current_percept = CHILD_NODE(temp_node, action).state
            print_state(current_percept)
            step += 1
    else:
        print("Don't have solution!")
```
