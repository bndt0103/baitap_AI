```python
#Ma trận mxn, giá trị 0 -> không bụi, giá trị 1 -> có bụi,
              # giá trị 2 -> vị trí robot
import random

floor = [
    [1, 1, 1],
    [1, 1, 1],
    [1, 1, 1]
]

row = len(floor)
col = len(floor[0])

vacuum_x = random.randint(0, row - 1)
vacuum_y = random.randint(0, col - 1)
```

```python
def show_floor():
    for i in range(row):
        for j in range(col):
            # Hiển thị máy hút bụi
            if i == vacuum_x and j == vacuum_y:
                print(2, end=" ")

            else:
                print(floor[i][j], end=" ")

        print()
```

```python
def has_dirty():
    for row in floor:
        if 1 in row:
            return True
    return False
```

```python
def rule_match():
    global vacuum_x, vacuum_y
    state = floor[vacuum_x][vacuum_y]

    print("Vị trí máy:", (vacuum_x, vacuum_y))
    if state == 0:
        print("Ô hiện tại = 0 (Clean)")
        directions = []
        if vacuum_x > 0:
            directions.append((-1, 0, "UP"))

        if vacuum_x < row - 1:
            directions.append((1, 0, "DOWN"))
            
        if vacuum_y > 0:
            directions.append((0, -1, "LEFT"))

        if vacuum_y < col - 1:
            directions.append((0, 1, "RIGHT"))

        move = random.choice(directions)
        dx, dy, direction_name = move
        vacuum_x += dx
        vacuum_y += dy
        print("ACTION: Move", direction_name)


    elif state == 1:
        print("Ô hiện tại = 1 (Dirty)")
        floor[vacuum_x][vacuum_y] = 0
        print("ACTION: Hút")
```

```python

step = 1
while has_dirty():
    print(f"\n========== STEP {step} ==========")
    show_floor()
    rule_match()
    step += 1
show_floor()
```
