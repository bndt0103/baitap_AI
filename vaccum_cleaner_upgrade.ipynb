```python
import random

# Ma trận mxn: 
# 0 -> không bụi
# 1 -> có bụi
# -1 -> vật cản (MỚI)
# 2 -> vị trí robot (hiển thị khi in)

floor = [
    [1, -1, 1],
    [1,  1, 1],
    [-1, 0, 1]
]

row = len(floor)
col = len(floor[0])

# Khởi tạo vị trí ngẫu nhiên, đảm bảo không rơi vào ô có vật cản
while True:
    vacuum_x = random.randint(0, row - 1)
    vacuum_y = random.randint(0, col - 1)
    if floor[vacuum_x][vacuum_y] != -1:
        break
```

```python
def show_floor():
    for i in range(row):
        for j in range(col):
            # Hiển thị máy hút bụi
            if i == vacuum_x and j == vacuum_y:
                print(2, end=" ")
            # Hiển thị vật cản
            elif floor[i][j] == -1:
                print("X", end=" ")
            else:
                print(floor[i][j], end=" ")
        print()
```

```python
def has_dirty():
    for row in floor:
        if 1 in row:
            return True
    return False
```

```python
def rule_match():
    global vacuum_x, vacuum_y
    state = floor[vacuum_x][vacuum_y]

    print("Vị trí máy:", (vacuum_x, vacuum_y))
    if state == 0:
        print("Ô hiện tại = 0 (Clean)")
        directions = []
        
        # Thêm điều kiện: Không vượt quá biên VÀ ô tiếp theo không phải là vật cản (-1)
        if vacuum_x > 0 and floor[vacuum_x - 1][vacuum_y] != -1:
            directions.append((-1, 0, "UP"))

        if vacuum_x < row - 1 and floor[vacuum_x + 1][vacuum_y] != -1:
            directions.append((1, 0, "DOWN"))
            
        if vacuum_y > 0 and floor[vacuum_x][vacuum_y - 1] != -1:
            directions.append((0, -1, "LEFT"))

        if vacuum_y < col - 1 and floor[vacuum_x][vacuum_y + 1] != -1:
            directions.append((0, 1, "RIGHT"))

        # Chọn hướng đi ngẫu nhiên từ các hướng hợp lệ
        if directions:
            move = random.choice(directions)
            dx, dy, direction_name = move
            vacuum_x += dx
            vacuum_y += dy
            print("ACTION: Move", direction_name)
        else:
            print("ACTION: Đứng yên (Bị kẹt bởi vật cản)")

    elif state == 1:
        print("Ô hiện tại = 1 (Dirty)")
        floor[vacuum_x][vacuum_y] = 0
        print("ACTION: Hút")
```

```python
step = 1
max_steps = 200 # Giới hạn số bước để tránh lặp vô hạn nếu rác không thể với tới

while has_dirty() and step <= max_steps:
    print(f"\n========== STEP {step} ==========")
    show_floor()
    rule_match()
    step += 1

print("\n========== KẾT QUẢ ==========")
if not has_dirty():
    print("Đã hoàn thành! Toàn bộ sàn đã sạch.")
else:
    print("Dừng thuật toán: Vượt quá số bước cho phép hoặc rác bị chặn bởi vật cản.")
show_floor()
```
