```python
import random

# Ma trận mxn: 
# 0 -> không bụi
# 1 -> có bụi
# -1 -> vật cản (MỚI)
# 2 -> vị trí robot (hiển thị khi in)

floor = [
    [0, 1, -1],
    [1, -1, 0],
    [0, 1, 1]
]
row = len(floor)
col = len(floor[0])

while True:
    vaccum_x = random.randint(0, row - 1)
    vaccum_y = random.randint(0, col - 1)
    if floor[vaccum_x][vaccum_y] != -1:
        break
visited = set()
path_stack = []
```

```python
def show_floor():
    for i in range(row):
        for j in range(col):
            if i == vaccum_x and j == vaccum_y:
                print('V', end=' ')
            elif floor[i][j] == -1:
                print('X', end=' ')
            else:
                print(floor[i][j], end=' ')
        print()
```

```python
def has_dirty():
    for row in floor:
        if 1 in row:
            return True
    return False
def update_state():
    visited.add((vaccum_x, vaccum_y))
```

```python
def rule_match():
    global vaccum_x, vaccum_y, path_stack
    visited.add((vaccum_x, vaccum_y))
    state = floor[vaccum_x][vaccum_y]
    
    if state == 1:
        print("Place is dirty!")
        floor[vaccum_x][vaccum_y] = 0
        print("Action: Suck!")
        if not has_dirty():
            print("All of floor is cleaned, done!")
            return False
    else:
        print("Place is cleaned!")

    direction = []
    if vaccum_x > 0 and floor[vaccum_x - 1][vaccum_y] != -1:
        direction.append((-1, 0, "UP"))
    if vaccum_x < row - 1 and floor[vaccum_x + 1][vaccum_y] != -1:
        direction.append((1, 0, "DOWN"))
    if vaccum_y > 0 and floor[vaccum_x][vaccum_y - 1] != -1:
        direction.append((0, -1, "LEFT"))
    if vaccum_y < col - 1 and floor[vaccum_x][vaccum_y + 1] != -1:
        direction.append((0, 1, "RIGHT"))

    unvisited = []
    for dx, dy, action in direction:
        nx = vaccum_x + dx
        ny = vaccum_y + dy
        if (nx, ny) not in visited:
            unvisited.append((dx, dy, action, nx, ny))

    if unvisited:
        move = random.choice(unvisited)
        dx, dy, direction_name, nx, ny = move
        path_stack.append((vaccum_x, vaccum_y))
        vaccum_x, vaccum_y = nx, ny
        print("Action: Move", direction_name)
        
    elif path_stack:
        px, py = path_stack.pop()
        if px < vaccum_x: dir_name = "UP"
        elif px > vaccum_x: dir_name = "DOWN"
        elif py < vaccum_y: dir_name = "LEFT"
        else: dir_name = "RIGHT"
        vaccum_x, vaccum_y = px, py
        print("Action: Backtrack ", dir_name)
    else:
        print("The robot has cleaned the entire accessible area, stop!")
        return False 
    return True
```

```python
step = 1
while True:
    print(f"\n===================== Step {step} ==================")
    show_floor()
    if not rule_match():
        break
    step += 1
```
