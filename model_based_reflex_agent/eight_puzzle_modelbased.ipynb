```python
import random

GOAL = [1, 2, 3, 4, 5, 6, 7, 8, 0]
```

```python
RULES = [
    ('UP',    -3),
    ('DOWN',  +3),
    ('LEFT',  -1),
    ('RIGHT', +1),
]

def check_condition(name, r, c):
    if name == 'UP':
        return r > 0      
    if name == 'DOWN':
        return r < 2      
    if name == 'LEFT':
        return c > 0      
    if name == 'RIGHT':
        return c < 2      
    return False
```

```python
def swap(state, i, j):
    new_state = state[:]     
    new_state[i] = state[j]
    new_state[j] = state[i]
    return new_state

def print_board(state):
    print(" ", state[0:3])
    print(" ", state[3:6])
    print(" ", state[6:9])
    print("-" * 15)
```

```python
def update_state(state, action, percept, model):
    model.append(percept) 
    return percept           
```

```python
def rule_match(state, rules, model):
    pos = state.index(0)     # Tìm vị trí ô trống
    r   = pos // 3           # Hàng của ô trống
    c   = pos % 3            # Cột của ô trống
    valid_actions = []       # Danh sách hành động hợp lệ
    for name, step in rules:
        if check_condition(name, r, c) == False:
            continue
        new_pos   = pos + step
        new_state = swap(state, pos, new_pos)
        if new_state not in model:
            valid_actions.append((name, new_state))
    if len(valid_actions) == 0:
        return None, None
    chosen = random.choice(valid_actions)
    return chosen[0], chosen[1]
```

```python
def model_based_reflex_agent(percept, state, model, action):
    state          = update_state(state, action, percept, model)
    action, next_s = rule_match(state, RULES, model)
    return action, next_s, state
```

```python
def run(init):
    state  = init[:]
    model  = []
    action = None
    print("TRẠNG THÁI BAN ĐẦU:")
    print_board(state)
    step = 1
    while True:                          
        if state == GOAL:
            print("THÀNH CÔNG sau", step - 1, "bước.")
            return
        action, next_s, state = model_based_reflex_agent(state, state, model, action)
        if action is None:             
            print("KẸT ở bước", step, ": Không còn đường đi mới.")
            return
        print("Bước", step, "Action move:", action)
        state = next_s
        print_board(state)
        step = step + 1                 
```

```python
init = [1, 2, 3,
        4, 6, 5,
        7, 8, 0]

run(init)
```
